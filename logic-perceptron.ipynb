```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
def and_gate(X1,X2): 
    w1 = w2 = 1.0
    b = -1.5

    z = (X1*w1)+ (X2*w2) + b
    return step(z)
```

```python
def step(z):
    return 1 if z>= 0 else 0
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(f"{x1}   {x2}  |   {and_gate(x1, x2)}")
```

```python
def or_gate(X1,X2): 
    w1 = w2 = 1.5
    b = -1.5
    z = (X1*w1)+ (X2*w2) + b
    return step(z)
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(f"{x1}   {x2}  |   {or_gate(x1, x2)}")
```

```python
def Nor_gate(X1,X2): 
    w1 = w2 = -1.5
    b = 1.0
    z = (X1*w1)+ (X2*w2) + b
    return step(z)
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(f"{x1}   {x2}  |   {Nor_gate(x1, x2)}")
```

```python
def nand_gate(X1,X2): 
    w1 = w2 = -1
    b = 1.0
    z = (X1*w1)+ (X2*w2) + b
    return step(z)
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(f"{x1}   {x2}  |   {nand_gate(x1, x2)}")
```
